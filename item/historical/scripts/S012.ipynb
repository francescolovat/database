# Getting the ISO code for each country

## Determining which countries do not have an ISO code
    Rule: As seen below, 16 countries do not have an ISO code. The following transformation is going to be used for each country name:
    
    ORIGINAL NAME --> NEW NAME
    Venezuela (Bolivarian Republic of) --> Venezuela
    Holy See --> Holy See (Vatican City State)
    China, Taiwan Province of China --> Taiwan
    State of Palestine --> Palestine
    Wallis and Futuna Islands -->  Wallis and Futuna
    Saint Helena --> Saint Helena, Ascension and Tristan da Cunha
    United States Virgin Islands --> Virgin Islands, U.S.
    Iran (Islamic Republic of) --> Iran, Islamic Republic of
    Dem. People's Republic of Korea --> Korea, Democratic People's Republic of
    Channel Islands --> ??
    Democratic Republic of the Congo --> Congo, The Democratic Republic of the
    China, Macao SAR --> Macao
    Bolivia (Plurinational State of) --> Bolivia
    Republic of Korea --> Korea, Republic of
    China, Hong Kong SAR --> Hong Kong
    Micronesia (Fed. States of) --> Micronesia, Federated States of

```python
list_of_countries = list(set(df[ColumnName.COUNTRY.value]))
countries_with_no_iso_code = countryCodeManager.get_list_of_countries_with_no_iso_code(list_of_countries)
countries_with_no_iso_code
```

```python
# Create a list to hold the new names of the countries
clean_list_of_country_names = []

# Getting a cleaned list of country names
for country in list(df["Country"]):
    if country == "Venezuela (Bolivarian Republic of)":
        clean_list_of_country_names.append("Venezuela")
    elif country == "Holy See":
        clean_list_of_country_names.append("Holy See (Vatican City State)")
    elif country == "China, Taiwan Province of China": 
        clean_list_of_country_names.append("Taiwan")
    elif country == "State of Palestine":
        clean_list_of_country_names.append("Palestine")
    elif country == "Wallis and Futuna Islands":
        clean_list_of_country_names.append("Wallis and Futuna")
    elif country == "Saint Helena":
        clean_list_of_country_names.append("Saint Helena, Ascension and Tristan da Cunha")
    elif country == "United States Virgin Islands":
        clean_list_of_country_names.append("Virgin Islands, U.S.")
    elif country == "Iran (Islamic Republic of)":
        clean_list_of_country_names.append("Iran, Islamic Republic of")
    elif country == "Dem. People's Republic of Korea":
        clean_list_of_country_names.append("Korea, Democratic People's Republic of")
    elif country == "Channel Islands":
        clean_list_of_country_names.append("Channel Islands")
    elif country == "Democratic Republic of the Congo":
        clean_list_of_country_names.append("Congo, The Democratic Republic of the")
    elif country == "China, Macao SAR":
        clean_list_of_country_names.append("Macao")
    elif country == "Bolivia (Plurinational State of)":
        clean_list_of_country_names.append("Bolivia")
    elif country == "Republic of Korea":
        clean_list_of_country_names.append("Korea, Republic of")
    elif country == "China, Hong Kong SAR":
        clean_list_of_country_names.append("Hong Kong")
    elif country == "Micronesia (Fed. States of)":
        clean_list_of_country_names.append("Micronesia, Federated States of")
    else:
        clean_list_of_country_names.append(country)

# Asserting that the new list is the correct size
assert len(clean_list_of_country_names) == len(df)

# Getting the list of iso codes
list_of_iso_codes = countryCodeManager.get_list_of_iso_for_countries(clean_list_of_country_names)

# Setting the new country column
df.drop(columns=["Country"], inplace = True)
df[ColumnName.COUNTRY.value] = clean_list_of_country_names

# Adding the column to the dataframe
df[ColumnName.ISO_CODE.value] = list_of_iso_codes
df
```

# Assigning the ITEM region

```python
# Getting the complete list of iso codes
list_of_all_codes = df[ColumnName.ISO_CODE.value]

# Getting the list of regions
item_regions = countryCodeManager.get_list_of_regions_for_iso_codes(list_of_all_codes)

# Assert that the size of the list of iso codes is equivalent to the size of country column
assert len(df) == len(item_regions)

# Adding the column to the dataframe
df.insert(3, ColumnName.ITEM_REGION.value, item_regions, True)
df
```

# Setting the "Value" column
    Rule: Currently the value column has the strings, we need to change it to a numeric value

```python
# List holding the values as numeric
list_with_values_as_numeric = []

# Getting all the values as numeric
for value in list(df["Value"]):
    new_value = value.replace(" ", "").replace(",", "")
    list_with_values_as_numeric.append(float(new_value))
    
# Asserting that the new list is of the correct length
assert len(list_with_values_as_numeric) == len(df)


# Removing the old "value" column
df.drop(columns=["Value"], inplace = True)

# Adding the new "value" column
df.insert(3, ColumnName.VALUE.value, list_with_values_as_numeric, True)
df
```
